# 📉 POLYDIM V4.2 - TEST DE ABLACIÓN Y GRÁFICAS
Este cuaderno genera las gráficas comparativas entre Polydim y el Baseline leyendo los logs de entrenamiento.

```python
# ============================================================
# POLYDIM V4.2 - ABLATION TEST & VISUALIZACIÓN
# ============================================================
# Este cuaderno está diseñado didácticamente para demostrar la 
# viabilidad de la topología Polydim frente a Transformers estándar.
# Carga los resultados del entrenamiento y grafica las curvas de convergencia.

import json
import matplotlib.pyplot as plt
import os

LOG_FILE = 'training_logs.json'

if not os.path.exists(LOG_FILE):
    print(f"❌ No se encontró {LOG_FILE}. Asegúrate de haber ejecutado primero 03_POLYDIM_V4_TinyShakespeare.ipynb")
else:
    print(f"✅ Cargando datos empíricos desde {LOG_FILE}...")
    with open(LOG_FILE, 'r') as f:
        history = json.load(f)

    # -----------------------------------------------------------
    # 1. CURVAS DE LOSS (PÉRDIDA)
    # -----------------------------------------------------------
    plt.figure(figsize=(10, 6))
    
    # Baseline
    plt.plot(history['baseline']['iters'], history['baseline']['train_loss'], 
             label='Baseline (Train)', color='red', alpha=0.3, linestyle='--')
    plt.plot(history['baseline']['iters'], history['baseline']['val_loss'], 
             label='Baseline (Val)', color='red', linewidth=2)
             
    # Polydim
    plt.plot(history['polydim']['iters'], history['polydim']['train_loss'], 
             label='Polydim V4.2 (Train)', color='blue', alpha=0.3, linestyle='--')
    plt.plot(history['polydim']['iters'], history['polydim']['val_loss'], 
             label='Polydim V4.2 (Val)', color='blue', linewidth=2)
             
    plt.title('Comparativa de Convergencia: Polydim vs Baseline Transformer', fontsize=14, pad=15)
    plt.xlabel('Iteraciones', fontsize=12)
    plt.ylabel('Cross-Entropy Loss', fontsize=12)
    plt.grid(True, alpha=0.3)
    plt.legend(fontsize=10)
    plt.tight_layout()
    plt.savefig('ablation_loss.png', dpi=300)
    plt.show()

    # -----------------------------------------------------------
    # 2. CURVAS DE ACCURACY (PRECISIÓN)
    # -----------------------------------------------------------
    plt.figure(figsize=(10, 6))
    
    # Baseline
    plt.plot(history['baseline']['iters'], history['baseline']['train_acc'], 
             label='Baseline (Train)', color='orange', alpha=0.3, linestyle='--')
    plt.plot(history['baseline']['iters'], history['baseline']['val_acc'], 
             label='Baseline (Val)', color='orange', linewidth=2)
             
    # Polydim
    plt.plot(history['polydim']['iters'], history['polydim']['train_acc'], 
             label='Polydim V4.2 (Train)', color='cyan', alpha=0.3, linestyle='--')
    plt.plot(history['polydim']['iters'], history['polydim']['val_acc'], 
             label='Polydim V4.2 (Val)', color='cyan', linewidth=2)
             
    plt.title('Precisión (Top-1 Acc): Polydim vs Baseline Transformer', fontsize=14, pad=15)
    plt.xlabel('Iteraciones', fontsize=12)
    plt.ylabel('Accuracy', fontsize=12)
    plt.grid(True, alpha=0.3)
    plt.legend(fontsize=10)
    plt.tight_layout()
    plt.savefig('ablation_acc.png', dpi=300)
    plt.show()
    
    print("\n📊 Gráficas guardadas como 'ablation_loss.png' y 'ablation_acc.png'.")

```
